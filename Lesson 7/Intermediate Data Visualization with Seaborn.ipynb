## 1. Seaborn Introduction

## 2.Customizing Seaborn Plots
Using Seaborn Styles

```python
us_housing['fmr_2'].plot.hist()
plt.show()
plt.clf()

sns.set()
us_housing['fmr_2'].plot.hist()
plt.show()
plt.clf()
```

```python
for style in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    sns.set_style(style)
    sns.displot(us_housing["fmr_2"] , kde=True, stat="density", kde_kws=dict(cut=3))
    plt.show()
    plt.clf()
```

```python
sns.set_style("white")
sns.lmplot(data=us_housing, x="pop2010", y="fmr_2")
sns.despine(left=True)
plt.show()
plt.clf()
```

Colors in Seaborn

```python
sns.set(color_codes=True)
sns.displot(us_housing["fmr_3"] , kde=True, stat="density", kde_kws=dict(cut=3), color='m')
plt.show()
```

```python
for p in ["bright", 'colorblind']:
    sns.set_palette(p)
    sns.displot(us_housing["fmr_3"] , kde=True, stat="density", kde_kws=dict(cut=3))
    plt.show()
    plt.clf()
```

```python
sns.palplot(sns.color_palette("Purples", 8))
plt.show()
```

Customizing with matplotlib

```python
import warnings

fig, ax = plt.subplots()
sns.distplot(us_housing["fmr_3"], kde=True, ax=ax)
ax.set(xlabel="3 Bedroom Fair Market Rent")

warnings.filterwarnings("ignore", category=UserWarning)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
sns.distplot(us_housing["fmr_1"], kde=True, ax=ax)
ax.set(xlabel="1 Bedroom Fair Market Rent", ylabel="Density", xlim=(100,1500), title="US Rent")
plt.show()
```

```python
fix, ax = plt.subplots()
sns.distplot(us_housing["fmr_1"], kde=True, ax=ax)
ax.set(xlabel="1 Bedroom Fair Market Rent", ylabel="Density", xlim=(100,1500), title="US Rent")

ax.axvline(us_housing["fmr_1"].median(), color="m", label="Median" , linestyle="--", linewidth=2)
ax.axvline(us_housing["fmr_1"].mean(), color="r", label="Mean" , linestyle="-", linewidth=2)

ax.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
sns.distplot(us_housing["fmr_1"], kde=True, ax=ax1)
ax1.set(xlabel="1 Bedroom Fair Market Rent", ylabel="Density", xlim=(100,1500))

sns.distplot(us_housing["fmr_2"], kde=True, ax=ax2)
ax2.set(xlabel="2 Bedroom Fair Market Rent", ylabel="Density", xlim=(100,1500))

plt.show()
```

## 4. Creating Plots on Data Aware Grids
Using FacetGrid, factorplot and lmplot

```python
g2 = sns.FacetGrid(college, col="Degree_Type", row_order=["Graduate", 'Bachelors', 'Associates', 'Certificate'])

g2.map(sns.boxplot, "SAT_AVG_ALL")
plt.show()
plt.clf()
```

```python
g2 = sns.FacetGrid(college, col="Degree_Type", row_order=["Graduate", 'Bachelors', 'Associates', 'Certificate'])
g2.map(sns.pointplot, "SAT_AVG_ALL")
plt.show()
plt.clf()
```

```python
sns.catplot(data=college, x="Tuition", kind="box", row="Degree_Type")
plt.show()
plt.clf()
```

```python
sns.catplot(data=college, x="SAT_AVG_ALL", kind="point", row="Degree_Type", row_order=["Graduate", 'Bachelors', 'Associates', 'Certificate'])
plt.show()
plt.clf()
```

```python
degree_ord = ['Graduate', 'Bachelors', 'Associates']
g = sns.FacetGrid(college, col="Degree_Type", col_order=degree_ord)
g.map(plt.scatter, 'UG', 'PCTPELL')
```

```python
sns.lmplot(data=college, x="UG", y="PCTPELL", col="Degree_Type", col_order=degree_ord)
plt.show()
plt.clf()
```

```python
inst_ord = ["Public", "Private non-profit"]
sns.lmplot(
    data=college,
    x="SAT_AVG_ALL",
    y="Tuition",
    col="Ownership",
    col_order=inst_ord,
    hue="WOMENONLY",
    row="Degree_Type",
    row_order=["Graduate", "Bachelors"],
)
plt.show()
plt.clf()
```

Using PairGrid and pairplot

```python
g = sns.PairGrid(us_worst_drives, vars=["fatal_collisions", "premiums"])
g2 = g.map(plt.scatter)
plt.show()
plt.clf()
```

```python
g = sns.PairGrid(us_worst_drives, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(plt.hist)
g3 = g2.map_offdiag(plt.scatter)
plt.show()
plt.clf()
```

```python
sns.pairplot(us_worst_drives, vars=["fatal_collisions", "premiums"], kind="scatter")
plt.show()
plt.clf()
```

```python
sns.pairplot(
    us_worst_drives,
    vars=["fatal_collisions", "premiums"],
    kind="scatter",
    hue="Region",
    diag_kws={"alpha": 0.5},
    palette="RdBu",
)
plt.show()
plt.clf()
```

```python
sns.pairplot(
    us_worst_drives,
    x_vars=["fatal_collisions_speeding", "fatal_collisions_alc"],
    y_vars=["premiums", "insurance_losses"],
    kind="scatter",
    hue="Region",
    palette="husl",
)
plt.show()
plt.clf()
```

```python
sns.pairplot(
    us_worst_drives,
    vars=["fatal_collisions", "premiums"],
    kind="reg",
    hue="Region",
    palette="BrBG",
    diag_kind="kde",
)
plt.show()
plt.clf()
```

Using JointGrid and jointplot

```python
sns.set_style("whitegrid")
g = sns.JointGrid(x="hum", y="total_rentals", data=bike_share, xlim=(0.1,1.0))
g.plot(sns.regplot, sns.distplot)
plt.show()
plt.clf()
```

```python
sns.jointplot(x='hum', y='total_rentals', data=bike_share, kind='reg')
plt.show()
plt.clf()
```

```python
sns.jointplot(
    x="temp", y="total_rentals", data=bike_share, kind="reg", xlim=(0, 1), order=2
)
plt.show()
plt.clf()
```

```python
sns.jointplot(x="temp", y="total_rentals", data=bike_share, kind="resid", order=2)
plt.show()
plt.clf()
```

```python
g = sns.jointplot(
    x="temp",
    y="casual",
    data=bike_share,
    kind="scatter",
    marginal_kws=dict(bins=10, rug=True),
).plot_joint(sns.kdeplot, cmap="mako")
plt.show()
plt.clf()
```

```python
g = sns.jointplot(
    x="temp",
    y="registered",
    data=bike_share,
    kind="scatter",
    marginal_kws=dict(bins=10, rug=True),
).plot_joint(sns.kdeplot, cmap="mako")
plt.show()
plt.clf()
```
