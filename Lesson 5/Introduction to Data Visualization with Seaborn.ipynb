## Using pandas with Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/Pokemon.csv")
sns.countplot(x="Total", data=df)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', style='smoker', hue_order=['Yes', 'No'])
plt.show()
```

```python
hue_colors = {'Yes': '#ff7777', 'No': '#F9FA9B'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', palette=hue_colors)
plt.show()
```

```python
sns.countplot(x='smoker', data=tips, hue='sex')
plt.show()
```

# 2.Visualizing Two Quantitative Variables
Introduction to relational plots and subplots

```python
# relplot() --> Relational plot; create subplots in a single figure

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', row='time', col='smoker')
plt.show()
```

```python
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    kind="scatter",
    col="day",
    col_wrap=2,
    col_order=["Thur", "Fri", "Sat", "Sun"],
)
plt.show()
```

## Customizing scatter plots

```python
tips.head(3)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, size='size', alpha=0.4)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, size='size', hue='size')
plt.show()
```

## Introduction to line plots

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
sns.relplot(x='model_year', y='mpg', data=mpg, kind='scatter')
```

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line")
plt.show()
```

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line",ci="sd")
plt.show()
```

```python
sns.relplot(
    x="model_year",
    y="horsepower",
    data=mpg,
    kind="line",
    ci=None,
    style="origin",
    hue="origin",
    markers=True,
    dashes=False,
)
plt.show()
```

# 3. Visualizing a Categorical and a Quantitative Variable
Count plots and bar plots

```python
category_order = ["Sun", "Saturday", "Fri", "Thur"]
sns.catplot(x="day", data=tips, kind="count", order=category_order)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="bar")
plt.show()
```

```python
sns.catplot(y="day", x="total_bill", data=tips, kind="bar")

plt.show()
```

```python
sns.catplot(y="day", x="total_bill", data=tips, kind="bar", col="smoker")

plt.show()
```

## Box plots

```python
sns.catplot(x="time", y="total_bill", data=tips, kind="box", order=["Dinner", "Lunch"])
plt.show()
```

```python
sns.catplot(
    x="time",
    y="total_bill",
    data=tips,
    kind="box",
    order=["Dinner", "Lunch"],
    whis=[0, 99],
)
plt.show()
```

## Pont plots

```python
sns.catplot(x="day", y="total_bill", data=tips, hue="smoker", kind="bar")
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", data=tips, hue="smoker", kind="point", join=False)
plt.show()
```

```python
from numpy import median

sns.catplot(
    x="smoker", y="total_bill", data=tips, kind="point", capsize=0.2, estimator=median
)
plt.show()
```

# 4. Customizing Seaborn Plots
Changing plot style and color

```python
sns.set_style("whitegrid")
sns.set_palette("RdBu")
sns.set_context("talk")
sns.catplot(
    x="smoker", y="total_bill", data=tips, kind="point", capsize=0.2, estimator=median
)
plt.show()
```

## Adding titles and labels

```python
g = sns.catplot(y="day", x="total_bill", data=tips, kind="bar", col="smoker")
g.fig.suptitle("Total Bill by Day and Smoker")
g.set_titles("Smoker: {col_name}")
plt.tight_layout()
plt.show()
```

## Putting it all together

```python
survey_data = pd.read_csv("../data/young-people-survey-responses.csv")
survey_data.head(3)
```
